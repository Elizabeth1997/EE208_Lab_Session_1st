```python
# plot
def figure(x_data, y_data,save_path=None, ran=[445, 670, 0, 20000]):
    fig, ax = plt.subplots()
    ax.plot(x_data, y_data)
    ax.set(xlabel='Wavelength (nm)', ylabel='Intensity (counts)')
    plt.axis(ran)
    if save_path:
        fig.savefig('{}.png'.format(save_path))
        print('The plot has been saved.')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
def fit_linear(x_data, y_data, save_path=None):
    gradient, intercept, r_value, p_value, std_err = stats.linregress(x_data,y_data)
    mn=np.min(x_data)
    mx=np.max(x_data)
    x1=np.linspace(mn,mx,500)
    y1=gradient*x1+intercept
    plt.plot(x_data,y_data,'ob')
    # plot the fitting curve
    plt.plot(x1,y1,'-r')
    plt.xlabel('The percentage of the power')
    plt.ylabel('FWHM')
    if save_path:
        fig.savefig('{}.png'.format(save_path))
        print('The linear fitting curve has been saved.')
    plt.show()
```
